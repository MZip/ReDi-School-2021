```python
# Now let's get started with our analysis. For that we will have a look at our target column first. Pandas allows us to
# analyse single columns in more detail by specifying the column name e.g. 'target' in squared brackets []. We will see the
# index of our dataframe and the values of the column (in this case zeros and ones)
df["target"]
```

```python
# Let's look at how the values of our target change if we look at the characteristics of one variable in more details
df.groupby('target')["cp"].value_counts().unstack(0).plot.bar(figsize=(15,5))
```

```python
# Split the data into input (X) and output (y)
y = df['target']
X = df.drop(columns=['target'])
```

```python
# Import a helper function to split the data into training and test
from sklearn.model_selection import train_test_split
```

```python
# Actually splitting the data based on a defined split value and after shuffeling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle=True)
```

```python
# Import a machine learning model for classification tasks from the sklearn libary - we are using a decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
# Creating an instance of the machine learning model after defining the parameter
decision_tree_model = DecisionTreeClassifier(criterion = 'entropy', max_depth=3, random_state = 33)
```

```python
# Training the machine learning model on training data
decision_tree_model.fit(X_train, y_train)
```

```python
# Making a prediction on test data
y_predict = decision_tree_model.predict(X_test)
```

## Evaluation

```python
# Load the confusion matrix function for our evaluation
from sklearn.metrics import plot_confusion_matrix
```

```python
# Plotting the confusion matrix with our test data
plot_confusion_matrix(decision_tree_model, X_test, y_test)
```

```python
# Loading a function to calculating the accuracy of our model
from sklearn.metrics import accuracy_score
```

```python
# Acutall calculating the accuracy
accuracy_score(y_test, y_predict)
```
